```python



# # txt파일을 코랩에서 작동하기 위해 업로드 한 코드 ---
# from google.colab import files

# # 파일 업로드
# uploaded = files.upload()

# print(uploaded)
# # -----------------------------------------------------

import re
from collections import Counter  # Counter 모듈 임포트

# 파일 경로 설정
file_path = './Avengers.txt'  # 파일의 경로를 지정합니다.

# 1. 텍스트 파일을 읽기 모드로 엽니다.
with open(file_path, 'r', encoding='utf-8') as file:  # 'r' 모드는 읽기 전용입니다.
    # 2. 파일의 내용을 읽어옵니다.
    content = file.read()  # 전체 내용을 읽습니다.

# 4. 내용을 소문자로 변환합니다.
lowercase_content = content.lower()

# 5. 모든 기호를 제거합니다.
# 정규 표현식 r'[^\w\s]'는 알파벳, 숫자, 공백을 제외한 모든 문자를 찾습니다.
cleaned_content = re.sub(r'[^\w\s]', '', lowercase_content)
## re.sub 를 써야해서 re 모듈을 import 해야된다는 것을 파악함

# 5. 단어 단위로 분리합니다.
words = cleaned_content.split()

# 6. 각 단어의 빈도수를 계산하여 딕셔너리 형태로 저장합니다.
word_counts = Counter(words)


## Counter 모듈이 있어야 실행 되는 것을 확인
## 기본적으로 내림차순 되어 있는 것 같음

# 7. 빈도수 순으로 내림차순 정렬합니다.
sorted_word_counts = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)

# 8. 정수 인덱스를 부여한 결과를 딕셔너리 형태로 저장합니다.
indexed_word_counts = {word : index for index, (word, count) in enumerate(sorted_word_counts)}

# # 10. 입력을 받아서 인덱스에 따른 빈도수를 반환하는 함수 정의
# def get_frequency_from_input():
#     input_text = input("문장을 입력하세요: ")  # 사용자로부터 입력 받기
#     input_words = input_text.lower().split()  # 입력된 문장을 소문자로 변환하고 단어 단위로 분리
#     frequencies = []  # 결과 빈도수를 저장할 리스트

#     for word in input_words:
#         # 각 단어에 대한 빈도수를 가져오고, 없다면 0
#         frequency = word_counts.get(word, 0)
#         frequencies.append(frequency)  # 빈도수를 리스트에 추가

#     return frequencies  # 빈도수 리스트 반환

# # 11. 함수 호출 및 결과 출력
# frequency_result = get_frequency_from_input()  # 입력한 단어에 대한 빈도수 가져오기
# print(f"입력한 단어의 빈도수: {frequency_result}")
## 우리가 원한건 딕셔너리의 밸류값인데, 이 코드는 인덱스의 빈도수가 출력되게끔 작성되서, 의도와 달라서 주석처리함


# # 9. 결과 출력, 테스트시 확인용으로 사용하였음
# print("단어 인덱스:", indexed_word_counts)  # 정수 인덱스와 단어 출력

# 10. 입력을 받아서 인덱스에 따른 단어를 반환하는 함수 정의
def get_index_from_input():
    input_text = input("문장을 입력하세요: ")  # 사용자로부터 입력 받기
    input_words = input_text.lower().split()  # 입력된 문장을 소문자로 변환하고 단어 단위로 분리
    indices = []  # 결과 인덱스를 저장할 리스트

    for word in input_words:
        # 각 단어에 대한 인덱스를 가져오고, 없다면 -1 또는 다른 기본값 설정
        index = indexed_word_counts.get(word, -1)  # 인덱스가 없으면 -1 반환
        indices.append(index)  # 인덱스를 리스트에 추가

    return indices  # 인덱스 리스트 반환

# 11. 함수 호출 및 결과 출력
index_result = get_index_from_input()  # 입력한 단어에 대한 인덱스 가져오기
print(f"입력한 단어의 인덱스: {index_result}")

## 회고
## 지금까지 했던 학습내용들을 숙지하고 있지 못 한것 같아 많이 아쉽다. 또한
## 또 대부분을 구글링과 ChatGPT에 의존한 것 같다. 대신 그러한 코드들은 찾아보고 공부를 더 해보겠다.
## 같이 소통 하면서 프로그래밍 해보니, 내가 잘못된 점이나 미처 생각하지 못했던 점, 다른관점 등을 파악할 수 있어서 좋았다.
```

```python
import re
from collections import Counter

def process_text_from_file(file_path):
    # 1. Avengers.txt 파일 불러오기
    with open(file_path, "r", encoding="utf-8") as file:
        avengers_text = file.read()  # 파일 내용 전체를 읽어옴

    # 2. 모든 문자를 소문자로 변환
    cleaned_text = avengers_text.lower()

    # 3. 모든 기호를 제거
    cleaned_text = re.sub(r'[^\w\s]', '', cleaned_text)

    # 4. 단어 단위로 분리
    words = cleaned_text.split()

    # 5. 각 단어 별 빈도수를 딕셔너리 형태로 저장
    word_counts = Counter(words)

    # 6. 딕셔너리를 빈도수 순으로 내림차순 정렬하고, 정수 인덱스를 부여
    sorted_word_counts = dict(sorted(word_counts.items(), key=lambda item: item[1], reverse=True))
    indexed_word_counts = {index: word for index, word in enumerate(sorted_word_counts.keys())}
    frequency_counts = {word: count for word, count in sorted_word_counts.items()}

    return indexed_word_counts, frequency_counts

def get_frequencies_from_input(user_input, indexed_word_counts, frequency_counts):
    # 입력된 문장을 소문자로 변환하고 기호 제거
    cleaned_input = re.sub(r'[^\w\s]', '', user_input.lower())

    # 단어 단위로 분리
    input_words = cleaned_input.split()

    # 각 단어의 빈도수를 정수 인덱스에 맞게 리스트로 생성
    frequency_list = [frequency_counts.get(word, 0) for word in input_words]

    return frequency_list

# 파일 경로 설정
file_path = "/content/Avengers.txt"

# 파일 처리
indexed_word_counts, frequency_counts = process_text_from_file(file_path)

# 사용자 입력 받기
user_input = input("텍스트를 입력하세요: ")

# 입력된 텍스트에 대해 빈도수 계산
frequencies = get_frequencies_from_input(user_input, indexed_word_counts, frequency_counts)

# 결과 출력
print(frequencies)
```
