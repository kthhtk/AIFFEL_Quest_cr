```python
import math

while True:
    #     
    while True:
        try:
            num1 = float(input("   : "))
            break
        except ValueError:
            print(" .  .")

    #     
    while True:
        try:
            num2 = float(input("   : "))
            break
        except ValueError:
            print(" .  .")

    #   
    while True:
        operator = input("  (+, -, *, /, **): ")
        if operator in ['+', '-', '*', '/', '**']:
            break
        print("  .  .")

    #  
    try:
        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                print("0   .")
                continue
            result = num1 / num2
        elif operator == '**':
            result = math.pow(num1, num2)

        print(f" : {num1} {operator} {num2} = {result}")

    except Exception as e:
        print(f"   : {e}")
        continue

    #  
    if input(" ? (y  ): ").lower() != 'y':
        print(" .")
        break
```
