```python
import math

while True:
    # 첫 번째 숫자 입력 받기
    while True:
        try:
            num1 = float(input("첫 번째 숫자를 입력하세요: "))
            break
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

    # 두 번째 숫자 입력 받기
    while True:
        try:
            num2 = float(input("두 번째 숫자를 입력하세요: "))
            break
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

    # 연산자 입력 받기
    while True:
        operator = input("연산자를 입력하세요 (+, -, *, /, **): ")
        if operator in ['+', '-', '*', '/', '**']:
            break
        print("지원하지 않는 연산자입니다. 다시 입력해주세요.")

    # 계산 수행
    try:
        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                print("0으로 나눌 수 없습니다.")
                continue
            result = num1 / num2
        elif operator == '**':
            result = math.pow(num1, num2)

        print(f"계산 결과: {num1} {operator} {num2} = {result}")

    except Exception as e:
        print(f"계산 중 오류가 발생했습니다: {e}")
        continue

    # 계속할지 확인
    if input("계속 계산하시겠습니까? (y 입력시 계속): ").lower() != 'y':
        print("계산기를 종료합니다.")
        break
```
