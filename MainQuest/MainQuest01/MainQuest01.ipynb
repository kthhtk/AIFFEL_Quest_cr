```python
import random

class Account:
    account_count = 0
    customer_list = []

    def __init__(self, name, balance):
        Account.account_count += 1
        Account.customer_list.append(self)
        self.bank = 'SC 은행'
        self.account_number = self.generate_account_number()
        self.name = name
        self.balance = balance
        self.transaction_count = 0  # 거래 카운트
        self.deposit_log = []
        self.withdraw_log = []
        self.deposit_count = 0  # 입금 횟수만 별도로 카운트

    def generate_account_number(self):
        r_num = random.randint
        return f'{r_num(0, 999):03d}-{r_num(0, 99):02d}-{r_num(0,999999):06d}'

    def get_account_num(self):
        print(f'총 계좌수 : {Account.account_count}')

    def deposit(self, money):
        if money <= 0:
            print('1원 이상 입금 해주세요')
        else:
            self.balance += money
            self.transaction_count += 1
            self.deposit_count += 1
            # 입금 내역 추가 (회차, 거래 유형, 금액, 잔액)
            self.deposit_log.append((self.transaction_count, '입금', money, self.balance))
            print(f'{money:,} 원이 입금되었습니다. 현재 잔액은 {self.balance:,} 원입니다.')
            self.interest()

    def withdraw(self, money):
        if money > self.balance:
            print(f'잔액이 부족합니다. 현재 잔액은 {self.balance:,} 원입니다.')
        else:
            self.balance -= money
            self.transaction_count += 1
            # 출금 내역 추가 (회차, 거래 유형, 금액, 잔액)
            self.withdraw_log.append((self.transaction_count, '출금', money, self.balance))
            print(f'{money:,} 원이 출금되었습니다. 현재 잔액은 {self.balance:,} 원입니다.')

    def display_info(self):
        customer_number = Account.customer_list.index(self) + 1
        return f'{customer_number}번 고객 - 은행: {self.bank}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원'

    def interest(self):
        if self.deposit_count % 5 == 0 and self.deposit_count > 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            self.transaction_count += 1
            # 이자 지급을 입금 내역에 추가
            self.deposit_log.append((self.transaction_count, '이자', interest, self.balance))
            print(f'{interest:,} 원의 이자가 지급되었습니다. 현재 잔액은 {self.balance:,} 원입니다.')

    @classmethod
    def list_100(cls):
        count = 0
        for customer in cls.customer_list:
            if customer.balance >= 1000000:
                print(customer.display_info())
                count += 1
        print(f' = 100만원 이상 고객 목록 = : 총 {cls.account_count} 명 중, {count} 명')

    def deposit_history(self):
        print(f"{self.name}님의 입금 내역:")
        for record in self.deposit_log:
            print(f"{record[0]}회 {record[1]} : {record[2]:,} 원, 잔액: {record[3]:,} 원")

    def withdraw_history(self):
        print(f"{self.name}님의 출금 내역:")
        for record in self.withdraw_log:
            print(f"{record[0]}회 {record[1]} : {record[2]:,} 원, 잔액: {record[3]:,} 원")

# 테스트 코드
a = Account('김태훈', 1000000)
b = Account('대훈', 2000000)
c = Account('소훈', 500000)

a.get_account_num()     # 총 계좌수 호출

# a 고객 입금 테스트
a.deposit(1000000)
a.deposit(1000000)
a.deposit(1000000)
a.deposit(1000000)
a.deposit(1000000)
a.deposit(1000000)

# a 고객 출금 테스트
a.withdraw(1000000)
a.withdraw(1000000)
a.withdraw(1000000)

# 입금 및 출금 내역 출력 테스트
print()
a.deposit_history()
a.withdraw_history()

# 100만원 이상 고객 목록 출력 테스트
print()
a.list_100()
```
