```python
import random

# 01. 캐릭터 클래스
class Character:
    def __init__(self, name, lv, hp, att, dff):
        self.name = name
        self.lv = lv
        self.hp = hp
        self.att = att
        self.dff = dff

    # 생존 확인 메서드
    def is_alive(self):
        return self.hp > 0

    # 맞는 데미지
    def take_damage(self, damage):
        if damage - self.dff > 0:
            self.hp -= damage - self.dff
        else:
            print(f'{self.name}는(은) 공격을 방어했다!')  # 0데미지 메시지 추가

    # 공격 데미지
    def attack_target(self):
        return random.randint(1, self.att)

# 02. 플레이어 클래스
class Player(Character):
    def __init__(self, name):
        super().__init__(name, lv=1, hp=100, att=25, dff=5)
        self.exp = 0

    # 경험치 상승 메서드
    def gain_experience(self, exp):
        self.exp += exp
        self.level_up()

    # 레벨 업 메서드
    def level_up(self):
        if self.exp >= 50:
            self.lv += 1
            self.att += 10
            self.dff += 5
            self.exp = 0
            print(f'\n=== Level Up! ===')
            print(f'레벨이 {self.lv}이 되었습니다!')
            print(f'공격력 +10 (현재 공격력: {self.att})')
            print(f'방어력 +5 (현재 방어력: {self.dff})')
            print(f'==================')

# 03. 몬스터 클래스
class Monster(Character):
    def __init__(self, name, lv):
        hp = random.randint(10, 30) * lv
        att = random.randint(5, 20) * lv
        dff = random.randint(1, 5) * lv
        super().__init__(name, lv, hp, att, dff)

# 04. 배틀 함수
def battle(player, monster):
    print(f'\n[전투 시작] {player.name}(LV.{player.lv}) VS {monster.name}(LV.{monster.lv})')

    while player.is_alive() and monster.is_alive():
        # 플레이어 턴
        damage = player.attack_target()
        monster.take_damage(damage)
        if damage - monster.dff > 0:  # 데미지가 들어갔을 때만 출력
            print(f'\n{player.name}의 공격! {damage}의 데미지! ({monster.name} HP:{monster.hp})')

        if not monster.is_alive():
            exp_gain = monster.lv * 20
            print(f'{monster.name} 처치! (경험치 +{exp_gain})')
            player.gain_experience(exp_gain)
            return

        # 몬스터 턴
        damage = monster.attack_target()
        player.take_damage(damage)
        if damage - player.dff > 0:  # 데미지가 들어갔을 때만 출력
            print(f'{monster.name}의 공격! {damage}의 데미지! ({player.name} HP:{player.hp})')

        if not player.is_alive():
            print('전투 패배...')
            return

# 05. 메인 함수
def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    player = Player(input('플레이어의 이름을 입력하세요: '))

    for monster_name, monster_level in monster_dict.items():
        print(f'\n{monster_name}이(가) 나타났다!')
        monster = Monster(monster_name, monster_level)

        battle(player, monster)

        if not player.is_alive():
            print('\n게임오버')
            return

    print('\n모든 몬스터를 물리쳤습니다! 게임 클리어!')

if __name__ == '__main__':
    main()
```
